DATA CLEANING

handling negative values

```python
df_fact_sales.isna().sum()
```

handling negative values using absolute method


```python
df_fact_sales['Qty']=df_fact_sales['Qty'].abs()
df_fact_sales
```

```python
neg_qty=df_fact_sales['Qty']<0
neg_qty.value_counts()
```

```python
neg_amt=df_fact_sales['net_sales_amount']<0
neg_amt.value_counts()
```

```python
df_dim_customer.isna().sum()
```

handling null values

```python
df_dim_market.isna().sum()
```

replacing the null values with 'not available'


```python
df_dim_market.fillna('not available', inplace=True)
df_dim_market
```

```python
df_dim_product.isna().sum()
```

calculating outlier limit

```python
higher_limit=df_fact_sales['net_sales_amount'].mean()+3*df_fact_sales['net_sales_amount'].std()
higher_limit
```

```python
outlier=df_fact_sales['net_sales_amount']>higher_limit
outlier.value_counts()
```

removing outlier values

```python
df_fact_sales=df_fact_sales[df_fact_sales['net_sales_amount']<higher_limit]
df_fact_sales
```
